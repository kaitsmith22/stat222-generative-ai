```python
import matplotlib.pyplot as plt
import pandas as pd
import torch

def plot_dataframe(df, column1, column2, experiment,):
    """
    Plot two columns of a Pandas dataframe as different lines.
    Args:
        df (pd.DataFrame): The Pandas dataframe to plot.
        column1 (str): The name of the first column to plot.
        column2 (str): The name of the second column to plot.
    """
    col1 = [float(elem[7:-1]) for elem in df[column1]]
    col2 = [float(elem[7:-1]) for elem in df[column2]]
    plt.plot(col1, label=column1)
    plt.plot(col2, label=column2)
    plt.xlabel("Epoch")
    plt.ylabel("Mean Absolute Error")
    plt.title("Training Loss for " + experiment)
    plt.legend()
    plt.show()

dat = pd.read_csv("experiments/Fix_Mask_VAE/Fix_Mask_VAE_training_metrics.csv")

plot_dataframe(dat, "Train_Loss", "Val_Loss", "VAE with Fixed Mask")
```

```python
dat = pd.read_csv("experiments/Base_VAE_50/Base_VAE_50_training_metrics.csv")

plot_dataframe(dat, "Train_Loss", "Val_Loss", "VAE with Random Mask")
```

```python
import os
import random
from PIL import Image, ImageOps

num_sample = 10000

# create data structure to hold all images
images = np.zeros((num_sample, 64 * 64))

for i, file in enumerate(random.sample(os.listdir('data/celeba/img_align_celeba'), num_sample)):
     image = Image.open(os.path.join('data/celeba/img_align_celeba',file))

     # convert to gray scale
     image = ImageOps.grayscale(image)

     # resize image to the same size we use in training
     image = image.resize((64,64))

     image_array = np.array(image).ravel() # it looks like the scipy function expects unravelled images

     # add image to numpy array
     images[i,:] = image_array
```
